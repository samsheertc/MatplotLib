```python
import pandas as pd
import numpy as np
df = pd.read_csv('./../data/survey_results_public.csv', index_col='Respondent')
schema_df = pd.read_csv('./../data/survey_results_schema.csv', index_col='Column')
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
df
```

```python
new_df= df[['Age','LanguageWorkedWith','ConvertedComp']]
new_df
```

```python
age_grp = new_df.groupby(['Age'])
age_avg = age_grp['ConvertedComp'].mean()
age_avg.dropna(inplace=True)
age_avg = age_avg.astype(int)
age_avg
```

```python
filt = new_df['LanguageWorkedWith'].str.contains('Python')
filt.replace(np.nan,False,inplace=True)
python = new_df[filt].groupby(['Age'])
python_avg = python['ConvertedComp'].mean()

python_avg.dropna(inplace=True)
python_avg = python_avg.astype(int)
python_avg
```

```python
filt = new_df['LanguageWorkedWith'].str.contains('JavaScript')
filt.replace(np.nan,False,inplace=True)
java = new_df[filt].groupby(['Age'])
java_avg = java['ConvertedComp'].mean()

java_avg.dropna(inplace=True)
java_avg = java_avg.astype(int)
java_avg
```

```python
final_df = pd.concat([age_avg, python_avg, java_avg], axis='columns', sort=False)
final_df.columns = ['All_Devs','Python','JavaScript']
final_df.replace(np.nan,0, inplace=True)
final_df.reset_index(inplace=True)
final_df = final_df.astype(int)
final_df.to_csv('./../data/modified.csv')
final_df
```

```python
import pandas as pd
from matplotlib import pyplot as plt

#data = pd.read_csv('modified.csv')

filt = (final_df['Age'] >30) &  (final_df['Age'] < 40)
data = final_df[filt]

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')


#Fill color between dev_salaries and py_salaries
plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries > dev_salaries) , interpolate=True, color='green', alpha=0.25, label='Above Avg')
plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries <= dev_salaries), interpolate=True, color='red'  , alpha=0.25, label='Below Avg')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```
